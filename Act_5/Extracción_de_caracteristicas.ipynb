```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from funpymodeling.exploratory import freq_tbl
```

```python
df=pd.read_excel("microretailer_mit_lift_lab.xlsx")
```

```python
df.info()
```

```python
df.head()
```

```python
freq_tbl(df)
```

```python
t1=freq_tbl(df['214_customer_relationship_tools'])
t1
```

```python
filtro=t1[t1['frequency']>1]
filtro
```

```python
filtro_index=filtro.set_index('214_customer_relationship_tools')
filtro_index
```

```python
filtro_index.plot(kind="bar",width=1,figsize=(10,4))
plt.title("Herramientas de relación con los clientes")
plt.xlabel("Tipo de Herramienta")
plt.ylabel("Frecuencia")
```

```python
filtro_index.plot("frequency","cumulative_perc",kind="scatter")
```

```python
filtro_index.plot(kind="area",figsize=(10,4),alpha=0.5)
```

```python
filtro_index["frequency"].plot(kind="pie", figsize=(10, 5), shadow=True, autopct="%0.1f %%", colors=['red', 'green', 'blue', 'yellow',"grey"])
```

```python
filtro_index.plot.hexbin("frequency","cumulative_perc",figsize=(10,5))
```
